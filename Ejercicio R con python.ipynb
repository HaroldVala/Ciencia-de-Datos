Funcion Query

La función query en pandas es una forma conveniente de filtrar filas en un DataFrame utilizando expresiones lógicas. Te permite seleccionar filas que cumplan con ciertas condiciones sin necesidad de escribir expresiones booleanas largas.

Donde "condición" es una cadena que representa la expresión lógica que deseas aplicar al DataFrame df.
Ejemplo: Supongamos que tienes un DataFrame llamado mi_df con las siguientes columnas: “edad”, “ingresos” y “ciudad”. Para filtrar las filas donde la edad sea mayor o igual a 30 y la ciudad sea “Tokio”, puedes hacer lo siguiente:

Consideraciones:
Las columnas en la expresión deben estar entre comillas si contienen espacios o caracteres especiales.
Puedes usar operadores lógicos como and, or y not.
La función query devuelve un nuevo DataFrame con las filas que cumplen con la condición.


```python
hogares_df[((hogares_df["clase_hog"] == 1) | (hogares_df["clase_hog"] == 2)) & (hogares_df["sexo_jefe"] == 2)]
```
